```python
Ans 2 
import pandas as pd

def reindex_dataframe(df):
    # Create a new index starting from 1 and incrementing by 2
    new_index = pd.RangeIndex(start=1, stop=len(df)*2, step=2)
    
    # Assign the new index to the DataFrame
    df_reindexed = df.set_index(new_index)
    
    return df_reindexed

import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({'A': [10, 20, 30],
                   'B': [40, 50, 60],
                   'C': [70, 80, 90]})

# Call the reindex_dataframe function
df_reindexed = reindex_dataframe(df)

# Print the re-indexed DataFrame
print(df_reindexed)
```

```python
Ans 3 
import pandas as pd

def calculate_sum(df):
    # Get the first three values in the 'Values' column
    values = df['Values'].head(3)

    # Calculate the sum of the values
    sum_values = values.sum()

    # Print the sum
    print("Sum of the first three values:", sum_values)


import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})

# Call the calculate_sum function
calculate_sum(df)
```

```python
Ans 4 
import pandas as pd

def add_word_count_column(df):
    # Split the 'Text' column by whitespace to count words
    word_count = df['Text'].str.split().apply(len)
    
    # Create a new column 'Word_Count' in the DataFrame
    df['Word_Count'] = word_count
    
    return df


import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({'Text': ['Hello world', 'This is a sentence', 'Python is awesome']})

# Call the add_word_count_column function
df_with_word_count = add_word_count_column(df)

# Print the DataFrame with the new column
print(df_with_word_count)
```

```python
Ans 7 
import pandas as pd

def extract_username(df):
    # Extract the username from each email address
    df['Username'] = df['Email'].str.split('@').str[0]
    
    return df

import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({'Email': ['john.doe@example.com', 'jane.smith@example.com']})

# Call the extract_username function
df_with_username = extract_username(df)

# Print the DataFrame with the new column
print(df_with_username)
```

```python
Ans 8 
import pandas as pd

def select_rows(df):
    # Select rows based on conditions
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]
    
    return selected_rows

import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({'A': [3, 8, 6, 2, 9],
                   'B': [5, 2, 9, 3, 1],
                   'C': [1, 7, 4, 5, 2]})

# Call the select_rows function
selected_df = select_rows(df)

# Print the selected DataFrame
print(selected_df)
```
